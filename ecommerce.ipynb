```python
import pandas as pd

df = pd.read_csv("Ecommerce- Dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# change the column name form Upper with spaces to lower case with underscore (snake_casing)

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace( ' ' , '_')
```

```python
df.columns
```

```python
# checking duplicate rows

duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows - {duplicate_rows}")
```

```python
# checking for null values 

df.isnull().sum()
```

```python
# Droppping null value rows
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
# Standardising Date columns

date_cols = ['invoicedate', 'signup_date', 'last_login_date']

for col in date_cols:
    df[col] = pd.to_datetime(df[col], format='mixed')
```

```python
# Removing negative values in price and quantity

df = df[(df['quantity'] > 0) & (df['price'] > 0)]
```

```python
# Ensure churn flag is an integer (not float)

df['churn_flag'] = df['churn_flag'].astype(int)
```

```python
# chcking the sahpe of the dataset after cleaning & before adding new column

df.shape
```

```python
# Adding column of days a customer first signup to last login

df['tenure_days'] = (df['last_login_date'] - df['signup_date']).dt.days
```

```python
# Ensuring all tenure days are positive as last login date !< signup date

df = df[(df['tenure_days'] > 0)]
```

```python
from sqlalchemy import create_engine

#Connecting to PostgreSQL
username = "postgres"      # default user
password = "admin" # the password you set during installation
host = "localhost"         # if running locally
port = "5432"              # default PostgreSQL port
database = "Ecommerce"    # the database you created in pgAdmin

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

#Loading DataFrame into PostgreSQL
table_name = "churn_rate"   # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
